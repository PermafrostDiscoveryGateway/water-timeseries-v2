# Basic Dataset Usage

This example notebook shows you how to access and use datasets 
1. Annual JRC surface water data
2. Dynamic World - monthly aggregated data

### Imports

```python
from water_timeseries.dataset import DWDataset, JRCDataset
import xarray as xr
import geopandas as gpd
```

#### Loading downloaded datacube (see Notebook 01)
* open data as xarray datasets
* load dataset as DWDataset Object - here we use the provided test set (cloned repo required)
* The DWDataset class runs automatic checks while being loaded, it expects:
  * xr.dataset with
    * 2 dimensions: "date", "id_geohash"
    * variables: ['water', 'bare', 'snow_and_ice', 'trees', 'grass', 'flooded_vegetation', 'crops', 'shrub_and_scrub', 'built', 'area_data']
* The dataset runs internal sanity checks
* The dataset saves the xr.dataset internal as .ds and creates a normalized datacube (to lake area are fraction [0:1]) as .ds_normalized

```python
ds_xr = xr.open_dataset('../tests/data/lakes_dw_test.zarr')
dataset = DWDataset(ds_xr)
```

The dataset contains several types of information

```python
dataset.data_columns # check data columns
dataset.water_column # column name of water specific variable
dataset.dates_ # available dates
dataset.object_ids_ # available object_ids (id_geohash in our case)
```

#### Access datacube
* original data with area values in ha
* normalized data with areas values as fraction (0-1) of the lake polygon

```python
# loaded xr.dataset
dataset.ds
# normalized xr.dataset
dataset.ds_normalized
```

#### Plot timeseries
The dataset has 2 simple to use plotting mechanisms included
* static plots (matplotlib)
* dynamic plots (plotly)

both allow saving the plot to disk
* png, jpg for static
* html for dynamic

```python
# show valid and available ids
dataset.object_ids_
```

Static plot for lake *b7uefy0bvcrc*

```python
fig = dataset.plot_timeseries(id_geohash='b7uefy0bvcrc')
```

Dynamic plot for lake *b7uefy0bvcrc*

```python
dataset.plot_timeseries_interactive(id_geohash='b7uefy0bvcrc')
```

You can save your plot to disk

```python
out_file = 'lake_timeseries_dw.png'
fig = dataset.plot_timeseries(id_geohash='b7uefy0bvcrc', save_path=out_file)
```

```python
out_file = 'lake_timeseries_dw.html'
dataset.plot_timeseries_interactive(id_geohash='b7uefy0bvcrc', save_path=out_file)
```

### Example for JRC Dataset
* using the JRC dataset basically works the same way and accounts for differences in data
  * temporal spacing (yearly)
  * different variables

```python
# loading
ds_xr = xr.open_dataset('../tests/data/lakes_jrc_test.zarr')
dataset_jrc = JRCDataset(ds_xr)
```

Plot static and save file

```python
out_file = 'lake_timeseries_jrc.png'
fig = dataset_jrc.plot_timeseries(id_geohash='b7uefy0bvcrc', save_path=out_file)
```

Plot interactive and save file

```python
out_file = 'lake_timeseries_jrc.html'
dataset_jrc.plot_timeseries_interactive(id_geohash='b7uefy0bvcrc', save_path=out_file)
```
